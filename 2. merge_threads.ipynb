```python
import pandas as pd
import os
```

```python
def get_data(file_name):
    collected_path = 'thread/'
    folder_path = f'{collected_path}{file_name}/'  # Specify the folder path
    file_names = []
    moral_data = pd.DataFrame()
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            f_name = os.path.join(file)
            if f_name.endswith('.csv'):
                if not 'checkpoint' in f_name:
                    moral_data = moral_data.append(pd.read_csv(folder_path+f_name),ignore_index=True)
                    
    print('moral_data: ',len(moral_data))
    return moral_data
```

```python
file_path = 'data/'
for file_name in ['zhihu_answers','zhihu_comments','weibo_posts', 'weibo_comments']:
    moral_detection = get_data(file_name)
    display(moral_detection.head())
    data = pd.read_csv(file_path+file_name+'.csv')
    data = data.merge(moral_detection, how = 'left',on = 'ID')
    data.to_csv(f'data/{file_name}_moral.csv',index=False)
```
